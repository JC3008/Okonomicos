```python
%scala
val path_dim_ativosb3 = "silver_db.dim_ativosb3"
val path_fato = "silver_db.fato_b3"
```

```python
%scala
val SCALA_fato_b3 = spark.read.table(path_fato)
SCALA_fato_b3.createOrReplaceTempView("Temp_SCALA_fato_b3")


val SCALA_ativos_b3 = spark.read.table(path_dim_ativosb3)
SCALA_ativos_b3.createOrReplaceTempView("Temp_SCALA_ativosb3")
```

```python
from pyspark.sql import functions as F

SPARK_ativosb3 = sqlContext.sql("""
select row_number() over(order by Cod) as pk_ativosb3, * from  Temp_SCALA_ativosb3 
""")
SPARK_ativosb3 = SPARK_ativosb3.withColumn('Gold_Loaded_Date',F.current_timestamp())

SPARK_ativosb3.createOrReplaceTempView("Temp_SPARK_ativosb3")
# display(sqlContext.sql("select * from Temp_SCALA_ativosb3"))
```

```python
%scala

val dim_calendario = "dbfs:/mnt/gold/dim_calendario.csv"
val SCALA_dim_calendario = spark.read.option("header",true).csv(dim_calendario)
SCALA_dim_calendario.createOrReplaceTempView("TEMP_SCALA_dim_calendario")
```

```python

SPARK_DF_FATO_B3 = sqlContext.sql("""

select a.*,b.pk_ativosb3,c.pk_dim_calendario from Temp_SCALA_fato_b3 a
left join Temp_SPARK_ativosb3 b on a.CodigoDeListagem = b.Cod
left join TEMP_SCALA_dim_calendario c on left(a.inbound_date,10)  = c.Date_2

""")
```
